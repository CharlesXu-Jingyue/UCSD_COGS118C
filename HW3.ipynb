```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

```python
data=pd.read_csv('HW3_Cogs118C_eeg.csv')
data
```

```python
dft = np.fft.fft(data['EEG'])
dft
```

```python
power = np.abs(dft)**2
plt.plot(power)
plt.xlabel('Frequency samples (k)')
plt.ylabel('Power spectrum (µV^2)')
plt.grid()
plt.yscale('log')
plt.xlim([0,20])
```

```python
f,Pxx = welch(data['EEG'],fs=500,nperseg=500)
```

```python
plt.semilogy(f, Pxx, '-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power (µV^2)')
```
